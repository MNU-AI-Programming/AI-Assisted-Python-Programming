## 조건문

이 노트북에서는 `if`, `elif`, `else`를 사용하여 조건에 따라 다른 결과를 출력하는 프로그램을 실습합니다.


## 실습 1. if 문으로 합격 출력하기

점수가 60점 이상이면 합격을 출력합니다.


````python
score = 80

if score >= 60:
    print("합격입니다.")
````

## 실습 2. 비교 연산자 확인하기

비교 연산의 결과는 `True` 또는 `False`입니다.


````python
score = 75

print(score >= 60)
print(score < 60)
print(score == 75)
print(score != 75)
````

## 실습 3. if-else 사용하기

조건이 참일 때와 거짓일 때를 나누어 처리합니다.


````python
score = int(input("점수를 입력하세요: "))

if score >= 60:
    print("합격입니다.")
else:
    print("불합격입니다.")
````

## 실습 4. elif 사용하기

여러 조건을 차례로 검사합니다.


````python
score = int(input("점수를 입력하세요: "))

if score >= 90:
    print("A학점")
elif score >= 80:
    print("B학점")
elif score >= 70:
    print("C학점")
elif score >= 60:
    print("D학점")
else:
    print("F학점")
````

## 실습 5. 학점 계산기 만들기

점수를 입력받아 학점을 변수에 저장한 뒤 출력합니다.


````python
score = int(input("점수를 입력하세요: "))

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"점수는 {score}점이고, 학점은 {grade}입니다.")
````

## 실습 6. 조건 순서 오류 확인하기

아래 코드는 왜 잘못된 결과가 나올까요?


````python
score = 95

if score >= 60:
    print("D학점")
elif score >= 70:
    print("C학점")
elif score >= 80:
    print("B학점")
elif score >= 90:
    print("A학점")
````

## 실습 7. 로그인 프로그램 만들기

`and` 연산자를 사용하여 아이디와 비밀번호가 모두 맞는지 확인합니다.


````python
user_id = input("아이디를 입력하세요: ")
password = input("비밀번호를 입력하세요: ")

if user_id == "admin" and password == "1234":
    print("로그인 성공")
else:
    print("로그인 실패")
````

## 실습 8. or 연산자 사용하기

둘 중 하나라도 참이면 실행됩니다.


````python
menu = input("메뉴를 입력하세요: ")

if menu == "커피" or menu == "차":
    print("음료를 주문했습니다.")
else:
    print("음료가 아닙니다.")
````

## 실습 9. 나이 판별 프로그램

나이를 입력받아 성인과 미성년자를 구분합니다.


````python
age = int(input("나이를 입력하세요: "))

if age >= 19:
    print(f"{age}살은 성인입니다.")
else:
    print(f"{age}살은 미성년자입니다.")
````

## 실습 11. 조건 판단 프로그램 만들기

아래 예시는 온도에 따른 메시지 프로그램입니다. 자신의 주제로 바꾸어 수정해 보세요.


````python
temperature = float(input("현재 온도를 입력하세요: "))

if temperature >= 30:
    message = "덥습니다. 물을 자주 마시세요."
elif temperature >= 20:
    message = "날씨가 좋습니다. 산책하기 좋습니다."
elif temperature >= 10:
    message = "쌀쌀합니다. 겉옷을 챙기세요."
else:
    message = "춥습니다. 따뜻하게 입으세요."

print(f"현재 온도는 {temperature}도입니다.")
print(message)
````

## 정리

이번 실습에서 배운 내용:

- `if`는 조건이 참일 때 실행됩니다.
- `else`는 조건이 거짓일 때 실행됩니다.
- `elif`는 여러 조건을 차례로 검사합니다.
- 조건문에는 콜론과 들여쓰기가 필요합니다.
- `and`, `or`로 여러 조건을 함께 사용할 수 있습니다.

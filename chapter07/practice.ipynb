## 리스트와 파일 처리

이 노트북에서는 리스트를 만들고, 반복문으로 처리하고, 성적 데이터를 CSV 파일로 저장하고 읽는 방법을 실습합니다.


## 실습 1. 리스트 만들기

여러 개의 값을 리스트에 저장합니다.


````python
students = ["김민수", "이서연", "박지훈"]
scores = [90, 85, 70]

print(students)
print(scores)
````

## 실습 2. 리스트 값 읽기

인덱스를 사용하여 리스트의 값을 읽습니다. 인덱스는 0부터 시작합니다.


````python
students = ["김민수", "이서연", "박지훈"]

print(students[0])
print(students[1])
print(students[2])
````

## 실습 3. 리스트 값 수정하기

인덱스를 사용하여 리스트의 값을 수정합니다.


````python
students = ["김민수", "이서연", "박지훈"]

students[1] = "최하늘"

print(students)
````

## 실습 4. 리스트에 값 추가하기

`append()`를 사용하여 리스트에 값을 추가합니다.


````python
students = ["김민수", "이서연"]

students.append("박지훈")

print(students)
````

## 실습 5. 빈 리스트에서 시작하기

빈 리스트를 만들고 값을 하나씩 추가합니다.


````python
scores = []

scores.append(90)
scores.append(80)
scores.append(100)

print(scores)
````

## 실습 6. 반복문으로 리스트 출력하기

리스트 안의 값을 하나씩 출력합니다.


````python
students = ["김민수", "이서연", "박지훈"]

for name in students:
    print(name)
````

## 실습 7. 점수 합계와 평균 구하기

`sum()`과 `len()`을 사용하여 평균을 계산합니다.


````python
scores = [90, 85, 70, 100]

total = sum(scores)
average = total / len(scores)

print(f"합계: {total}")
print(f"평균: {average:.2f}")
````

## 실습 8. 성적 관리 프로그램 만들기

학생 이름과 점수를 입력받아 리스트에 저장합니다.


````python
names = []
scores = []

for i in range(3):
    name = input("학생 이름을 입력하세요: ")
    score = int(input("점수를 입력하세요: "))

    names.append(name)
    scores.append(score)

print("===== 성적 목록 =====")

for i in range(3):
    print(f"{names[i]}: {scores[i]}점")

average = sum(scores) / len(scores)
print(f"평균 점수: {average:.2f}")
````

## 실습 9. CSV 파일 저장하기

리스트에 저장된 이름과 점수를 CSV 파일로 저장합니다.


````python
names = ["김민수", "이서연", "박지훈"]
scores = [90, 85, 70]

with open("scores.csv", "w", encoding="utf-8") as file:
    file.write("이름,점수\n")

    for i in range(len(names)):
        file.write(f"{names[i]},{scores[i]}\n")

print("scores.csv 파일이 저장되었습니다.")
````

## 실습 10. CSV 파일 읽기

저장한 CSV 파일을 다시 읽어 확인합니다.


````python
with open("scores.csv", "r", encoding="utf-8") as file:
    content = file.read()

print(content)
````

## 실습 11. CSV 파일을 한 줄씩 읽기

`readlines()`와 `strip()`을 사용합니다.


````python
with open("scores.csv", "r", encoding="utf-8") as file:
    lines = file.readlines()

for line in lines:
    print(line.strip())
````

## 실습 14. 나만의 성적 관리 프로그램 만들기

아래 코드를 수정하여 학생 수를 바꾸거나 출력 형식을 개선해 보세요.


````python
names = []
scores = []

student_count = int(input("학생 수를 입력하세요: "))

for i in range(student_count):
    name = input(f"{i+1}번째 학생 이름: ")
    score = int(input(f"{i+1}번째 학생 점수: "))
    names.append(name)
    scores.append(score)

average = sum(scores) / len(scores)

print("===== 성적 결과 =====")
for i in range(len(names)):
    print(f"{names[i]}: {scores[i]}점")
print(f"평균 점수: {average:.2f}")

with open("my_scores.csv", "w", encoding="utf-8") as file:
    file.write("이름,점수\n")
    for i in range(len(names)):
        file.write(f"{names[i]},{scores[i]}\n")

print("my_scores.csv 파일 저장 완료")
````

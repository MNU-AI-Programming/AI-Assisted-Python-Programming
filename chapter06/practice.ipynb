## 함수와 AI 활용

이 노트북에서는 함수를 만들고, 매개변수와 return을 사용하며, AI를 활용한 디버깅과 리팩토링을 실습합니다.


## 실습 1. 간단한 함수 만들기

`def`를 사용하여 함수를 만들어 봅니다.


````python
def hello():
    print("안녕하세요")
    print("AI 활용 코딩 수업입니다")

hello()
````

## 실습 2. 함수 여러 번 호출하기

함수는 여러 번 호출할 수 있습니다.


````python
def hello():
    print("안녕하세요")

hello()
hello()
hello()
````

## 실습 3. 매개변수 사용하기

함수에 값을 전달해 봅니다.


````python
def hello(name):
    print(f"안녕하세요, {name}님!")

hello("홍길동")
hello("김민수")
````

## 실습 4. 여러 매개변수 사용하기

이름과 학과를 받아 자기소개를 출력합니다.


````python
def introduce(name, major):
    print(f"이름: {name}")
    print(f"학과: {major}")

introduce("홍길동", "교양학부")
````

## 실습 5. return 사용하기

함수의 결과를 돌려받아 변수에 저장합니다.


````python
def add(a, b):
    result = a + b
    return result

answer = add(3, 5)
print(answer)
````

## 실습 6. print()와 return 비교하기

두 방식의 차이를 확인합니다.


````python
def add_print(a, b):
    print(a + b)

def add_return(a, b):
    return a + b

add_print(3, 5)

result = add_return(3, 5)
print(result)
print(result * 2)
````

## 실습 7. 계산기 함수 만들기

사칙연산 함수를 만들어 봅니다.


````python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b

x = 10
y = 2

print(f"{x} + {y} = {add(x, y)}")
print(f"{x} - {y} = {subtract(x, y)}")
print(f"{x} * {y} = {multiply(x, y)}")
print(f"{x} / {y} = {divide(x, y)}")
````

## 실습 8. BMI 계산 함수 만들기

키와 몸무게를 받아 BMI를 계산하는 함수를 만듭니다.


````python
def calculate_bmi(height_cm, weight):
    height_m = height_cm / 100
    bmi = weight / (height_m * height_m)
    return bmi

height = float(input("키를 cm 단위로 입력하세요: "))
weight = float(input("몸무게를 kg 단위로 입력하세요: "))

bmi = calculate_bmi(height, weight)

print(f"당신의 BMI는 {bmi:.2f}입니다.")
````

## 실습 11. 반복 코드를 함수로 리팩토링하기

반복되는 총 금액 계산 코드를 함수로 바꾸어 봅니다.


````python
def calculate_total(price, quantity):
    return price * quantity

total1 = calculate_total(3000, 2)
total2 = calculate_total(5000, 3)

print(f"첫 번째 총 금액: {total1}원")
print(f"두 번째 총 금액: {total2}원")
````

## 실습 12. 나만의 함수 만들기

아래 예시는 원의 넓이를 계산하는 함수입니다. 다른 함수로 수정해 보세요.


````python
def circle_area(radius):
    area = 3.14 * radius * radius
    return area

r = float(input("원의 반지름을 입력하세요: "))
area = circle_area(r)

print(f"반지름이 {r}인 원의 넓이는 {area:.2f}입니다.")
````

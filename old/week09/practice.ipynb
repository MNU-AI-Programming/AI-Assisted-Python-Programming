## 기본 예제 1


````python
import numpy as np
scores=np.array([85,92,76,88,95])
print(scores.shape,scores.dtype,scores.mean(),scores.std())
print(scores[scores>=80])
````

## 기본 예제 2


````python
attendance=np.array([92,95,80,90,98])
assignment=np.array([88,94,78,86,96])
total=scores*0.6+attendance*0.2+assignment*0.2
print(np.round(total,2))
````

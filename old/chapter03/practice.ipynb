## 입력과 계산

이 노트북에서는 사용자 입력, 숫자 변환, 산술 연산자, BMI 계산기, 간단한 계산기를 실습합니다.


## 실습 1. 이름 입력받기

`input()` 함수로 사용자 입력을 받아 봅시다.


````python
name = input("이름을 입력하세요: ")
print("안녕하세요,", name, "님!")
````

## 실습 2. 입력값의 자료형 확인하기

`input()`으로 입력받은 값은 문자열입니다.


````python
age = input("나이를 입력하세요: ")
print(age)
print(type(age))
````

## 실습 3. 정수로 변환하기

`int()`를 사용하여 문자열을 정수로 변환합니다.


````python
age = int(input("나이를 입력하세요: "))
print(f"내년 나이는 {age + 1}살입니다.")
````

## 실습 4. 실수로 변환하기

`float()`을 사용하여 소수점이 있는 숫자로 변환합니다.


````python
height = float(input("키를 입력하세요: "))
print(f"입력한 키는 {height}입니다.")
print(type(height))
````

## 실습 5. 산술 연산자 실습

두 숫자를 사용하여 여러 계산을 해 봅시다.


````python
a = 10
b = 3

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b}")
print(f"{a} % {b} = {a % b}")
print(f"{a} ** {b} = {a ** b}")
````

## 실습 6. 소수점 자리 조절하기

f-string으로 소수점 자리를 조절합니다.


````python
result = 10 / 3
print(f"결과: {result}")
print(f"결과: {result:.2f}")
````

## 실습 7. BMI 계산기 만들기

키를 m 단위로 입력받아 BMI를 계산합니다.


````python
height = float(input("키를 m 단위로 입력하세요: "))
weight = float(input("몸무게를 kg 단위로 입력하세요: "))

bmi = weight / (height * height)

print(f"당신의 BMI는 {bmi:.2f}입니다.")
````

## 실습 8. cm 단위 BMI 계산기 만들기

키를 cm 단위로 입력받은 후 m 단위로 변환합니다.


````python
height_cm = float(input("키를 cm 단위로 입력하세요: "))
weight = float(input("몸무게를 kg 단위로 입력하세요: "))

height_m = height_cm / 100
bmi = weight / (height_m * height_m)

print(f"키: {height_cm}cm")
print(f"몸무게: {weight}kg")
print(f"BMI: {bmi:.2f}")
````

## 실습 9. 간단한 계산기 만들기

두 숫자를 입력받아 사칙연산 결과를 출력합니다.


````python
a = float(input("첫 번째 숫자를 입력하세요: "))
b = float(input("두 번째 숫자를 입력하세요: "))

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
````

## 실습 11. 생활 계산기 만들기

아래 예시는 여행 경비 계산기입니다. 자신의 주제로 바꾸어 수정해 보세요.


````python
transport = int(input("교통비를 입력하세요: "))
hotel = int(input("숙박비를 입력하세요: "))
food = int(input("식비를 입력하세요: "))
etc = int(input("기타 비용을 입력하세요: "))

total = transport + hotel + food + etc

print("===== 여행 경비 계산 결과 =====")
print(f"교통비: {transport}원")
print(f"숙박비: {hotel}원")
print(f"식비: {food}원")
print(f"기타 비용: {etc}원")
print(f"총 경비: {total}원")
````

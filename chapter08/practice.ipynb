## 3. 예시 프로젝트 1: 용돈 관리 프로그램

가장 기본적인 형태의 용돈 관리 프로그램입니다. 자신의 프로젝트에 맞게 수정하세요.


````python
income = int(input("이번 주 수입을 입력하세요: "))
expense = int(input("이번 주 지출을 입력하세요: "))

balance = income - expense

print("===== 용돈 관리 결과 =====")
print(f"수입: {income}원")
print(f"지출: {expense}원")
print(f"잔액: {balance}원")

if balance < 0:
    print("지출이 수입보다 많습니다. 다음 주에는 지출을 줄여 보세요.")
else:
    print("잘 관리하고 있습니다.")
````

## 4. 예시 프로젝트 2: TODO 리스트

할 일을 리스트에 저장하고 출력하는 프로그램입니다.


````python
todo_list = []

count = int(input("몇 개의 할 일을 입력할까요? "))

for i in range(count):
    todo = input(f"{i+1}번째 할 일을 입력하세요: ")
    todo_list.append(todo)

print("===== TODO 리스트 =====")

for i in range(len(todo_list)):
    print(f"{i+1}. {todo_list[i]}")
````

## 5. 예시 프로젝트 3: 성적 관리 프로그램

학생 이름과 점수를 입력받아 평균을 계산합니다.


````python
names = []
scores = []

student_count = int(input("학생 수를 입력하세요: "))

for i in range(student_count):
    name = input(f"{i+1}번째 학생 이름: ")
    score = int(input(f"{i+1}번째 학생 점수: "))
    names.append(name)
    scores.append(score)

average = sum(scores) / len(scores)
highest = max(scores)
lowest = min(scores)

print("===== 성적 결과 =====")
for i in range(len(names)):
    print(f"{names[i]}: {scores[i]}점")

print(f"평균 점수: {average:.2f}")
print(f"최고 점수: {highest}")
print(f"최저 점수: {lowest}")
````

## 6. CSV 저장 예시

프로젝트 결과를 CSV 파일로 저장하는 예시입니다.


````python
# 성적 관리 프로그램에서 만든 names와 scores 리스트가 있다고 가정합니다.

with open("project_result.csv", "w", encoding="utf-8") as file:
    file.write("이름,점수\n")

    for i in range(len(names)):
        file.write(f"{names[i]},{scores[i]}\n")

print("project_result.csv 파일이 저장되었습니다.")
````

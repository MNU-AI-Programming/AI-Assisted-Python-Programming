## 기본 예제 1


````python
import pandas as pd
url="https://raw.githubusercontent.com/MNU-AI-Programming/AI-Assisted-Python-Programming/main/datasets/scores.csv"
df=pd.read_csv(url)
display(df.head())
print(df.shape)
print(df.columns.tolist())
print(df.dtypes)
````

## 기본 예제 2


````python
summary=df["score"].agg(["count","mean","min","max"])
high=df.loc[df["score"]>=80,["name","major","score"]]
sorted_df=df.sort_values("score",ascending=False)
display(summary); display(high); display(sorted_df.head())
````

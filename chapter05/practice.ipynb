## 반복문

이 노트북에서는 `for`, `while`, `range()`, `break`를 사용하여 반복되는 프로그램을 실습합니다.


## 실습 1. 같은 문장 여러 번 출력하기

반복문 없이 같은 문장을 여러 번 출력해 봅니다.


````python
print("안녕하세요")
print("안녕하세요")
print("안녕하세요")
print("안녕하세요")
print("안녕하세요")
````

## 실습 2. for 반복문 사용하기

`for`와 `range()`를 사용하여 같은 문장을 반복 출력합니다.


````python
for i in range(5):
    print("안녕하세요")
````

## 실습 3. range()로 숫자 출력하기

0부터 4까지 출력합니다.


````python
for i in range(5):
    print(i)
````

## 실습 4. 1부터 10까지 출력하기

`range(1, 11)`을 사용합니다.


````python
for i in range(1, 11):
    print(i)
````

## 실습 5. 1부터 10까지 합계 구하기

반복문을 사용하여 합계를 구합니다.


````python
total = 0

for i in range(1, 11):
    total = total + i

print(f"1부터 10까지의 합계는 {total}입니다.")
````

## 실습 6. 구구단 2단 출력하기

반복문으로 구구단을 출력합니다.


````python
dan = 2

for i in range(1, 10):
    print(f"{dan} x {i} = {dan * i}")
````

## 실습 7. 입력받은 단의 구구단 출력하기

사용자가 원하는 단을 입력합니다.


````python
dan = int(input("출력할 단을 입력하세요: "))

print(f"===== {dan}단 =====")

for i in range(1, 10):
    print(f"{dan} x {i} = {dan * i}")
````

## 실습 8. while 반복문 사용하기

`while`은 조건이 참인 동안 반복합니다.


````python
count = 1

while count <= 5:
    print(count)
    count = count + 1
````

## 실습 9. 짝수 출력하기

반복문과 조건문을 함께 사용합니다.


````python
for i in range(1, 11):
    if i % 2 == 0:
        print(i)
````

## 실습 10. 홀수 출력하기

위 코드를 수정하여 홀수만 출력합니다.


````python
for i in range(1, 11):
    if i % 2 == 1:
        print(i)
````

## 실습 11. 숫자 맞추기 게임

사용자가 정답을 맞힐 때까지 반복합니다.


````python
answer = 7
guess = 0

while guess != answer:
    guess = int(input("숫자를 맞혀 보세요: "))

    if guess == answer:
        print("정답입니다!")
    else:
        print("틀렸습니다. 다시 시도해 보세요.")
````

## 실습 12. random을 사용한 숫자 맞추기 게임

무작위 정답을 생성합니다.


````python
import random

answer = random.randint(1, 10)
guess = 0

while guess != answer:
    guess = int(input("1부터 10 사이의 숫자를 입력하세요: "))

    if guess == answer:
        print("정답입니다!")
    elif guess < answer:
        print("더 큰 숫자입니다.")
    else:
        print("더 작은 숫자입니다.")
````

## 실습 13. break로 반복문 종료하기

사용자가 0을 입력하면 반복문을 종료합니다.


````python
while True:
    number = int(input("숫자를 입력하세요. 종료하려면 0 입력: "))

    if number == 0:
        print("프로그램을 종료합니다.")
        break

    print(f"입력한 숫자: {number}")
````

## 실습 15. 반복문 프로그램 만들기

아래 예시는 1부터 N까지 합계를 구하는 프로그램입니다. 자신의 주제로 수정해 보세요.


````python
n = int(input("어디까지 더할까요? "))
total = 0

for i in range(1, n + 1):
    total += i

print(f"1부터 {n}까지의 합계는 {total}입니다.")
````

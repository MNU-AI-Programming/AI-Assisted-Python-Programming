## 실습 1. 문장 출력하기

`print()` 함수를 사용하여 문장을 출력해 봅시다.


````python
print("안녕하세요")
print("AI 활용 코딩 수업입니다")
````

## 실습 2. 여러 줄 출력하기

여러 개의 `print()`를 사용하면 여러 줄을 출력할 수 있습니다.


````python
print("이름: 홍길동")
print("학과: 교양학부")
print("관심 분야: 인공지능")
````

## 실습 3. 변수 사용하기

변수는 데이터를 저장하는 이름표입니다.


````python
name = "홍길동"
major = "교양학부"

print(name)
print(major)
````

## 실습 4. 변수로 자기소개 출력하기

변수에 저장된 값을 사용하여 자기소개를 출력합니다.


````python
name = "홍길동"
major = "교양학부"
interest = "인공지능"

print("안녕하세요. 제 이름은", name, "입니다.")
print("저는", major, "소속입니다.")
print("저의 관심 분야는", interest, "입니다.")
````

## 실습 5. 자료형 확인하기

`type()` 함수를 사용하여 자료형을 확인합니다.


````python
age = 20
height = 175.5
name = "홍길동"
is_student = True

print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
````

## 실습 6. 숫자와 문자열 비교하기

숫자 `10`과 문자열 `"10"`의 차이를 확인합니다.


````python
print(10 + 5)
print("10" + "5")
````

## 실습 7. 문자열 연결하기

`+`를 사용하여 문자열을 연결할 수 있습니다.


````python
first = "AI"
second = "Coding"

print(first + second)
print(first + " " + second)
````

## 실습 8. f-string 사용하기

f-string을 사용하면 문자열 안에 변수 값을 쉽게 넣을 수 있습니다.


````python
name = "홍길동"
age = 20
major = "교양학부"

print(f"안녕하세요. 저는 {name}입니다.")
print(f"저는 {age}살입니다.")
print(f"저는 {major} 소속입니다.")
````

## 실습 10. 나의 프로필 카드 만들기

아래 코드를 자신의 정보에 맞게 수정해 보세요.


````python
name = "홍길동"
major = "교양학부"
age = 20
interest = "인공지능"
favorite = "음악"
goal = "AI를 활용해 간단한 프로그램 만들기"

print("===== 나의 프로필 카드 =====")
print(f"이름: {name}")
print(f"학과: {major}")
print(f"나이: {age}")
print(f"관심 분야: {interest}")
print(f"좋아하는 것: {favorite}")
print(f"이번 학기 목표: {goal}")

print("나이 변수의 자료형:", type(age))
````
